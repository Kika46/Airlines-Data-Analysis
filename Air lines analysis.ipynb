Q4. Does price varies with airlines?

```python
df.columns  
```

```python
df.groupby('airline')['price'].mean().sort_values(ascending=False)
```

```python
sns.catplot(x='airline', y='price', data=df, kind='bar', palette='viridis', order=df.groupby('airline')['price'].mean().sort_values(ascending=False).index,hue='class')
plt.title('Average Ticket Price by Airline')
plt.xlabel('Airline')  
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

Q5. Does ticket price change based on the departure time and arrival time?

```python
df.columns
```

```python
df.groupby('departure_time')['price'].mean().sort_values(ascending=False)
```

```python
df.groupby('arrival_time')['price'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.catplot(x='departure_time', y='price', data=df, kind='bar', palette='magma',order=df.groupby('departure_time')['price'].mean().sort_values(ascending=False).index)
plt.title('Average Ticket Price by Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()


```

```python
plt.figure(figsize=(12,6))
sns.catplot(x='arrival_time', y='price', data=df, kind='bar', palette='magma',order=df.groupby('arrival_time')['price'].mean().sort_values(ascending=False).index)
plt.title('Average Ticket Price by Arrival Time')
plt.xlabel('Arrival Time')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.relplot(x='arrival_time', y='price', data=df, palette='Set2',col='departure_time',kind = 'line')
```
